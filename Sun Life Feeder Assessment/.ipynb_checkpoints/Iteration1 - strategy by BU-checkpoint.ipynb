```python
import pandas as pd

df = pd.read_excel('Output0325.xlsx', index_col=0)

df
```

```python
# is_derived =  df['Reqcategory']=='IFRS 17'
# summary_derived = df[is_derived]
# undecided =  df['BUProposedStrategytoMeetRequirements']=='Undecided'
# summary_derived = summary_derived[undecided]
# summary_derived

summary_derived = df[(df['Reqcategory']=='IFRS 17') & (df['BUProposedStrategytoMeetRequirements']=='Undecided')]
summary_derived

summary = pd.DataFrame({'Feed Count': summary_derived.groupby(['BG', 'BU','PRKey'])['PRKey'].count()}).reset_index()
summary['Percentage'] = 100 - (summary['Feed Count'] / 17 * 100)
summary
```

```python
summary.to_excel("buStrat0326.xlsx")
```

```python
file_types = df.Typename.unique()

file_types
```

```python
attributes = df.FinancialReportingRequirements.unique()
attributes
attributes_count = {}
for i in attributes:
    name = df.loc[df['FinancialReportingRequirements'].isin([i])]
    attributes_count[i] = name.FinancialReportingRequirements.count()

attributes_count
```

```python
summary['Percentage'] = ""
for key in attributes_count:
    new = summary.loc[(summary['FinancialReportingRequirements'].isin([key]))]
    new['Percentage'] = new['Feed Count'] / attributes_count[key] * 100
    summary.update(new)

summary
```

```python
summary.to_excel("summary0325.xlsx")
```
