```python
import pandas as pd

df = pd.read_excel('Output0324_v1.xlsx', index_col=0)
df
```

```python
summary = pd.DataFrame({'Feed Count': df.groupby(['BG', 'BU', 'Reqcategory', 'Typename', 'FinancialReportingRequirements', 'BUProposedStrategytoMeetRequirements'])['BUProposedStrategytoMeetRequirements'].count()}).reset_index()
summary
```

```python
file_types = df.Typename.unique()

file_types
```

```python

# counts = df[df['Typename'].isin(["New GB EBF"])].PRKey.count()
# counts
attributes = df.FinancialReportingRequirements.unique()
attributes
attributes_count = {}
for j in file_types:
    nested = {}
    for i in attributes:
        name = df.loc[(df['Typename'].isin([j])) & (df['FinancialReportingRequirements'].isin([i]))]
        nested[i] = name.FinancialReportingRequirements.count()
    attributes_count[j] = nested

attributes_count
```

```python
summary['Percentage'] = ""
for file_type in attributes_count:
    for key in attributes_count[file_type]:
        new = summary.loc[(summary['Typename'].isin([file_type])) & (summary['FinancialReportingRequirements'].isin([key]))]
        new['Percentage'] = new['Feed Count'] / attributes_count[file_type][key] * 100
        summary.update(new)

summary
```

```python
summary.to_excel("summary.xlsx")
```
